```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
SV_Exec = numpy.poly1d(numpy.polyfit(df['Input Size'], df['SV Exec'], 3))


```

```python
SV_Mem = numpy.poly1d(numpy.polyfit(df['Input Size'], df['SV Mem'], 3))

myline = numpy.linspace(10, 54, 100)

plt.scatter(df['Input Size'], df['SV Mem'])
plt.plot(myline, SV_Mem(myline))
plt.show()
```
